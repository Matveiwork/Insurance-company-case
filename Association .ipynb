```python
import pandas as pd
import sys
if not sys.warnoptions:
    import warnings
    warnings.simplefilter("ignore")
import numpy as np
from efficient_apriori import apriori
```

```python
df = pd.read_excel('input_data-2.xlsx')
```

```python
df.head()
```

```python
df_categorial = df[['product_name','contract_status','currency_name','country','sex']]
```

```python
columns = list(df_categorial.columns)

transactions = []
for i in range(len(df_categorial)):
    values=tuple(list(t.strip() for t in df_categorial.iloc[i,:].values if (t is not None) and (not isinstance(t,float))))
    transactions.append(values)
```

```python
transactions
```

```python
items,rules = apriori(transactions,min_support=0.009,min_confidence=0.9)
```
