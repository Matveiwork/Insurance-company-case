```python
client = pd.read_excel('CASE_CLIENTS.xlsx')
```

```python
looses = pd.read_excel('CASE_LOSSES.xlsx')
```

```python
contracs = pd.read_excel('CASE_CONTRACTS.xlsx')
```

```python
group = pd.read_excel('experiment_group.xlsx')
```

```python
df = pd.merge(client,looses,left_on='client_id',right_on='client_id',how='inner')
```

```python
df = pd.merge(df,contracs,left_on='client_id',right_on='client_id',how='inner')
```

```python
df = pd.merge(df,group,left_on='client_id',right_on='client_id',how='inner')
```

```python
df
```

```python
df_rub = df.loc[df['currency_name'].isin(['Российский рубль'])]
```

```python
df_usd = df.loc[df['currency_name'].isin(['Доллар США'])]
```

```python
df_rub['loss_payout_amt'] = df_rub['loss_payout_amt'].apply(lambda x:x/87.8)
```

```python
df_rub['insurance_amount'] = df_rub['insurance_amount'].apply(lambda x:x/87.8)
```

```python
df_rub['price'] = df_rub['price'].apply(lambda x:x/87.8)
```

```python
df_final = pd.concat([df_rub,df_usd])
```

```python
df_final
```

```python
df_final['currency_name'] = df_final['currency_name'].apply(lambda x: 'Доллар США' )
```
